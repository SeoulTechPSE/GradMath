### 5. The Integral Calculus of Functions of a Complex Variable

Let $f(z)$ be a function of the complex variables $z =x +iy$, and let us consider a curve $\mathrm{C}$ in the complex plane, with end points $a$ and $b$. We suppose that the curve $\mathrm{C}$ is described by a parametric equation

> $z(t)=x(t)+iy(t), \;\;t_a \leq t \leq t_b$

We first subdivide the arc $ab$ into $n$ intervals by introducing the $n+1$ points

> $\displaystyle z_0, z_1, z_2, \cdots, z_{n-1}, z_n$

Next we refine further the subdivision of $\mathrm{C}$ by introducing an additional series of points $\xi_1, \xi_2,\cdots,\xi_n$ taken along $\mathrm{C}$ and such that $\xi_k$ lies between $z_{k-1}$ and $z_k$

><img src="figs/gmath1_04.jpg" width="250">

We now form the sum

>$I_n=\sum_{k=1}^n f(\xi_k) (z_k -z_{k-1})$

and take the limit $n \,\to\infty$ in such a manner that $|z_k -z_{k-1}| \,\to 0$ for all $k$. This limit, provided it exists and provided it is independent of the manner in which we have chosen the points $z_j$ and $\xi_j$, is called the **contour integral** of $f(z)$ along $\mathrm{C}$ and is written as

>$\displaystyle I = \int_\mathrm{C} f(z)\,dz$

Separating $f(z)$ and $z$ into their real and imaginary parts, we can also write the sum $I_n$ as

>$\displaystyle I_n=\sum_{k=1}^n \left\{ \left[u(\xi_k)(x_k -x_{k-1}) -v(\xi_k)(y_k -y_{k-1}) \right] 
  +i \left[v(\xi_k)(x_k -x_{k-1}) +u(\xi_k)(y_k -y_{k-1}) \right]\right\}$
  
The limiting transition implies $|x_k -x_{k-1}| \,\to 0$ and $|y_k -y_{k-1}| \,\to 0$ for all $k$, and the integral $I$ can be expressed in terms of real curvlinear integrals as

>$\displaystyle I=\int_\mathrm{C} udx -vdy +i \int_\mathrm{C} vdx +udy$

* The function $F(z)$ is called the primitive function of $f(z)$

    >$\displaystyle f(z) = \frac{dF(z)}{dz}$

    Then, for a $z$ situated on the contour
    
    >$\displaystyle f(z)\frac{dz}{dt}=\frac{dF(z)}{dz}\frac{dz}{dt}=\frac{dF[z(t)]}{dt}$
    
    we get
    
    >$\displaystyle \int_\mathrm{C} f(z)\,dz=\int_{t_a}^{t_b} \frac{dF[z(t)]}{dt} \,dt=F(b)-F(a)$
    
* The formal properties of the contour integral

    * $\displaystyle \int_\mathrm{C} af(z) \,dz = a\int_\mathrm{C} f(z) \,dz$
    
    * $\displaystyle \int_\mathrm{C} \left\{f(z) +g(z) \right\} \,dz = \int_\mathrm{C} f(z) \,dz +\int_\mathrm{C} g(z) \,dz$
    
    * If $c$ is a point that divides the arc $ab$ into two arcs $ac$ and $cb$, then
    
    > $\displaystyle \int_a^b f(z) \,dz = \int_a^c f(z) \,dz +\int_c^b f(z) \,dz$
    
    > $\displaystyle \int_a^b f(z) \,dz = -\int_b^a f(z) \,dz$   
    
    * $\displaystyle \int_a^b \left.\frac{df(z)}{dz}\,g(z) \,dz = f(z)g(z)\,\right|_a^b -\int_a^b f(z)\frac{dg(z)}{dz} \,dz$

### 6. The Darboux Inequality

It is very often useful to consider the upper bounds of certain contour integrals

>$\begin{align*}
 I_n &= \sum_{k=1}^n f(\xi_k) (z_k -z_{k-1})\\ 
 &\Downarrow \\ 
 |I_n| &\leq \sum_{k=1}^n \left| f(\xi_k) \right| \left| z_k -z_{k-1}\right|  
 \leq \mathrm{max}\, |f| \,\sum_{k=1}^n \left| z_k -z_{k-1}\right| \leq \mathrm{max}\, |f| \cdot L \\
 &\Downarrow {\scriptstyle \text{as } n\,\to\infty} \\
 \left| \int_\mathrm{C} f(z) \,dz\, \right| &\leq \mathrm{max}\, |f| \cdot L
\end{align*}$

This is called **Darboux's inequality**

#### 7. Some Definitions

* A function is **analytic** at a given point of the complex plane if there exists a neighborhood of this point such that the function is single-valued and differentiable at all points of this neighborhood

* The set of all points where a function is analytic is called the **domain of analyticity** of the function. If the domain of analyticity is the entire complex plane, the function is called an **entire** function

* A point where a function is analytic is a **regular point** of the function. If a function is not analytic at some point, the point is called a **singular point** of the function

* For example, the single-valued function $f(z)=1/z$ is differentiable everywhere except at $z=0$, since its derivative at that point is infinite. Such an exceptional point is called an **isolated singular point** of the function

### 8. Examples of Analytic Functions



#### 8.1 Polynomials

* An arbitrary polynomial

    >$\displaystyle p_n(z)=\sum_{k=0}^n a_k z^k$
    
    is an entire function
    
    >$\begin{align*}
 \frac{dp_n(z)}{dz} &= \lim_{\Delta z \to\, 0} \frac{1}{\Delta z} \sum_{k=0}^n a_k \left[ (z +\Delta z)^k -z^k \right]\\ 
 &\Downarrow \;\;{\scriptstyle (z +\Delta z)^k \,=\, z^k +kz^{k-1}\Delta z +\cdots +(\Delta z)^k} \\ 
 &= \sum_{k=0}^n a_k k z^{k-1}
\end{align*}$

#### 8.2 Power Series

* A power series with an infinite number of terms

    >$\displaystyle f(z)=\sum_{k=0}^\infty a_k (z -z_0)^k$
    
    converges for $|z -z_0| < R$, in which $R$ is called the **radius of convergence** of the corresponding power series. Any power series is an analytic function within its radius of convergence

#### 8.3 Exponential and Related Functions

* The exponential function is defined by its power series expansion

    >$\displaystyle e^z=\sum_{k=0}^\infty \frac{z^k}{k!}$
    
    This equation has an infinite radius of convergence, since for any $z$ one has
    
    >$\displaystyle \sum_{k=0}^\infty \frac{|z|^k}{k!}=e^{|z|}$
    
    which, by virtue of Weierstrass' criterion, implies the convergence of the sum. Therefore, $e^z$ is an entire function
    
* As in the real variable case, one can prove that

    >$\begin{align*}
 e^{z +w} &= \sum_{k=0}^\infty \frac{1}{k!} (z +w)^k \\
 &= \sum_{k=0}^\infty \frac{1}{k!} \sum_{l=0}^k \frac{k!}{l! (k-l)!} z^l w^{k-l}  \\ 
 &= \sum_{l=0}^\infty \frac{z^l}{l!} \sum_{k = l}^\infty \frac{w^{k-l}}{(k-l)!} \\
 &= e^z e^w
\end{align*}$

* When $y$ is a real variable,

    >$\begin{align*}
 e^{iy} &= \sum_{k=0}^\infty \frac{(iy)^k}{k!}\\ &=\left(1 -\frac{y^2}{2!} +\frac{y^4}{4!} +\cdots \right) +
    i\left(y -\frac{y^2}{3!} +\frac{y^5}{5!} +\cdots \right) \\ 
 &= \cos y +i\sin y\\ 
 &\Downarrow \\ 
 \cos y &= \frac{1}{2} \left( e^{iy} +e^{-iy} \right) \\
 \sin y &= \frac{1}{2i} \left( e^{iy} -e^{-iy} \right) 
\end{align*}$

* The definition of the trigonometric functions for complex arguments

    >$\begin{align*}
 \cos z &= \frac{1}{2} \left( e^{iz} +e^{-iz} \right) \\
 \sin z &= \frac{1}{2i} \left( e^{iz} -e^{-iz} \right) \\
 &\Downarrow \\
 \cos z &= \sum_{k=0}^\infty (-1)^k \frac{z^{2k}}{(2k)!} \\
 \sin z &= \sum_{k=0}^\infty (-1)^k \frac{z^{2k+1}}{(2k+1)!} \\
 &\Downarrow \\
 \sin^2 z &+\cos^2 z =1
\end{align*}$

* The definition of hyperbolic functions

    >$\begin{align*}
 \cosh z &= \cos iz \\
 \sinh z &= -i\sin iz \\
 &\Downarrow \\
 \cosh^2 z &-\sinh^2 z =1 \\
 \cosh z &= \frac{1}{2} \left( e^{z} +e^{-z} \right) \\
 \sinh z &= \frac{1}{2} \left( e^{z} -e^{-z} \right) \\
\end{align*}$

    The functions $\tan z$, $\tanh z$, $\sec z$, $\csc z$, etc, are defined as in the case of real vaiables
    
* The logarithm $\ln z$ of a complex variable

    >$\begin{align*}
 e^{f(z)} &= z \\
 &\Downarrow f(z) =u +i v \\
 e^u (\cos z +i \sin z) &= r (\cos\theta +i\sin\theta) \\
 &\Downarrow \\
 u &= \ln r \\
 v &= \theta +2n\pi, \;\;n=0,\pm1,\pm2, \cdots \\
 &\Downarrow \\
 \ln z &= \ln r +i(\theta +2n\pi), \;\;n=0,\pm1,\pm2, \cdots
\end{align*}$

    * The principal logarithm
    
    >$\displaystyle \mathrm{Ln} z = \ln r +i\theta, \;\;-\pi<\theta <\pi$
    
    $\mathrm{Ln} z$ has a discontinuity across the negative real axis
    
    >$\displaystyle \lim_{\epsilon\to\,0} \mathrm{Ln}(re^{i(\pi \pm \epsilon)}) =\ln r \pm i\pi$
    
    In the entire complex plane with the exception of the negative real axis, the principal logarithm is an analytic function

### 9. Conformal Transformations

#### 9.1 Conformal Mapping

* Consider a function

    >$\displaystyle z'(z)=x' +iy'=u(x,y) +iv(x,y)$ $\;\;(*)$

    of the complex variable $z=x+iy$, which is analytic in a region $\mathrm{D}$ of the complex $z$ plane. 
    
* Since the mapping is continuous, two continuous curves $\mathbf{C}_1$ and $\mathbf{C}_2$ in the $z$ plane will be mapped into two other continuous curves $\mathbf{C}'_1$ and $\mathbf{C}'_2$ in the $z'$ plane and the point of intersection $z_0=x_0+iy_0$ of $\mathbf{C}_1$ and $\mathbf{C}_2$ will go over into the point of intersection  $z'_0=x'_0+iy'_0$ of $\mathbf{C}'_1$ and $\mathbf{C}'_2$.
Let $\nu$ be the angle between the tangents to $\mathbf{C}_1$ and $\mathbf{C}_2$ at $z_0$, let $\nu'$ be the angle between the tangents to $\mathbf{C}'_1$ and $\mathbf{C}'_2$ at $z'_0$

    ><img src="figs/gmath1_05.jpg" width="350">

    For all points $z \in \mathrm{D}$ such that
    
    >$\displaystyle \frac{dz'(z)}{dz} \neq 0$
    
    the mapping $(*)$ is **angle-preserving**; $\nu=\nu'$
    
* Let $z$ be an arbitrary point on $\mathrm{C}_1$ and $z'$ the image of $z$ on $\mathrm{C}'_1$

    >$\begin{align*}
 z -z_0 &= re^{i\theta}\\ 
 z' -z'_0 &= r' e^{i\theta'},\;\; \frac{dz'(z_0)}{dz_0}= \left| \frac{dz'(z_0)}{dz_0} \right| e^{i\psi_0}\\
 &\Downarrow \\
 \frac{z' -z'_0}{z -z_0} &= \frac{r'}{r} e^{i(\theta' -\theta)} \\
 &\Downarrow {\scriptstyle z \to\, z_0 \text{ along } \mathrm{C}_1}\\
 \theta &\rightarrow \alpha_1, \;\; \theta' \rightarrow \beta_1 \Rightarrow \psi_0 =\beta_1 -\alpha_1 \\
 &\Downarrow {\scriptstyle z \to\, z_0 \text{ along } \mathrm{C}_2}\\
 \theta &\rightarrow \alpha_2, \;\; \theta' \rightarrow \beta_2 \Rightarrow \psi_0 =\beta_2 -\alpha_2 \\
  &\Downarrow \\
  \beta_1 -\alpha_1 &= \beta_2 -\alpha_2 \\
    &\Downarrow \\
  \nu = \alpha_2 -\alpha_1 &= \beta_2 -\beta_1 = \nu'
\end{align*}$

    The angle between the tangents at a point of intersection of any two regular curves $\mathrm{C}_1 \in \mathrm{D}$ and $\mathrm{C}_2 \in \mathrm{D}$ is preserved under the mapping $\nu=\nu'$ when 
    
    >$\displaystyle \frac{dz'}{dz}\neq 0 \;\;\text{for } z \in \mathrm{D}$
    
    Such a mapping is called a **conformal mapping**
    

#### 9.2 Homographic Transformations

* The transformation

    >$\displaystyle z' =\frac{1}{z}$
    
    is called an **inversion**. It affects a conformal mapping of any region in the $z$ plane that does not include the point $z=0$
    
    >$\begin{align*}
 |z -a| &= r \\ 
 &\Downarrow {\scriptstyle |a| \neq r}\\ 
 \left| \frac{1}{z'} -a \right| &= r \\
 &\Downarrow \\ 
 |1-az'| &= r|z'| \\
 &\Downarrow \\ 
 |z'|^2 -\frac{2}{|a|^2 -r^2} \mathrm{Re} (az') &+\frac{1}{|a|^2 -r^2} =0 \\
 &\Downarrow \\ 
 \left| z' -A \right| &=R
\end{align*}$

    where 
    
    >$\displaystyle A = \frac{a}{|a|^2 -r^2}$ and $\displaystyle R = \left| \frac{r}{|a|^2 -r^2} \right|$
    
    The inversion transformation maps circles in the z plane into circles in the $z'$ plane or straight lines
    
    >$2x' \mathrm{Re}\, a -2y' \mathrm{Im}\, a -1 = 0$
    
    in the case of $|a|=r$ in the $z'$ plane
    
    The image of the origin is defined to be *the point at infinity*. The image of every line that reaches the point $z=0$ is a line that extends to infinity. Similarly, one can speak of the point $z=\infty$ as the image of the point $z'=0$
    
* Consider the series of transformation

    >$\begin{align*}
 z_1 &= z +\frac{d}{c} \\ 
 z_2 &= c^2 z_1 \\ 
 z_3 &= \frac{1}{z_2}\\ 
 z_4 &=(bc -ad) z_3 \\ 
 z' &= \frac{a}{c} +z_4 
\end{align*}$

    where $c\neq 0$ and $bc-ad \neq 0$
    
    All these transformations map circles in the $z$ plane into circles (or straight lines) in the $z'$ plane. The successive application of these transformations
    
    >$\displaystyle z'=\frac{a}{c} +z_4=\frac{a}{c}+(bc -ad)z_3=\cdots=\frac{az +b}{cz +d}$
    
    also maps circles into circles (or straight lines) and is called a **homographic** transformation. Using the well-known rules of differentiation, one obtain
    
    >$\displaystyle \frac{dz'}{dz}=\frac{ad -bc}{(cz +d)^2}$
    
    The condition $c\neq 0$ and $bc-ad \neq 0$ ensure that $dz'/dz \neq 0$, and therefore a homographic transformation maps conformally every region in the $z$ plane that does not include the point $z=-(d/c)$. In the case where $c=0$, a homographic transformation reduces to a linear one

#### 9.3 Change of Integration Variable

* Consider the integral $\displaystyle\int_\mathrm{C} f(z)\, dz$ taken along a regular curve $\mathrm{C}$, with a parametric equation $z=z(t)$, $\;t_a \leq t \leq t_b$. A conformal mapping $z'=z'(z)$ establishes a one-to-one correspondence between points in the $z$ plane and their images in the $z'$ plane, and consequently there must exist a function $z=z(z')$ inverse to $z'=z'(z)$. Reducing contour integrals to Riemann integrals, we easily obtain

    >$\displaystyle \int_\mathrm{C} f(z)\,dz = \int_\mathrm{C} f[z(z')] \left(\frac{dz(z')}{dz'}\right) \,dz'$

### 10. A Simple Application of Conformal Mapping 

* Consider the following electrostatic problem: Given two parallel conducting cylinders of infinite length, what is the electric field at any point of space due to these cylinders if their surfaces are at given potentials?

    We shall solve the problem in the case when the diameters of the cylinders are equal. The problem is in fact a two-dimensional one, since the electric field cannot depend upon the coordinate directed along the axes of the cylinders; the field vector $\mathbf{E}$ depends on two Cartesian coordinates $x$ and $y$, in any plane perpendicular to the axes of the cylinders

* Let $U=U(x,y)$ be the electrostatic potential. In empty space, $U(x,y)$ satisfies Laplace's equation

    >$\displaystyle \frac{\partial^2 U(x,y)}{\partial x^2} +\frac{\partial^2 U(x,y)}{\partial y^2}=0$
    
    $U(x,y)$ determines $\mathbf{E}(x,y)$ by the well-known relation
    
    >$\mathbf{E}(x,y)=-\nabla U(x,y)$
    
    or in terms of components
    
    >$\begin{align*}
 E_x(x,y) &= -\frac{\partial U(x,y)}{\partial x}\\ 
 E_y(x,y) &= -\frac{\partial U(x,y)}{\partial y}
\end{align*}$
    
* Since $U(x,y)$ satisfies Laplace's equation, this suggests the introduction of another function $V(x,y)$ such that the pair of functions $U(x,y)$ and $V(x,y)$ satisfy the Cauchy-Riemann conditions

    >$\begin{align*}
 \frac{\partial V(x,y)}{\partial x} &= -\frac{\partial U(x,y)}{\partial y}\\ 
 \frac{\partial V(x,y)}{\partial y} &= \frac{\partial U(x,y)}{\partial x}
\end{align*}$
    
    It is easy to verify that in terms of $V(x,y)$, $\mathbf{E}(x,y)$ is given by
    
    >$\mathbf{E}(x,y)=-\nabla \times \mathbf{n} V(x,y)$
    
    where $\mathbf{n}$ is a unit normal to the $x$, $y$ plane
    
* $U(x,y)$ and $V(x,y)$ together define a function

    >$\displaystyle F(z)=U(x,y) +iV(x,y)$
    
    called the **complex potential**, which is analytic at all points of empty space where $U(x,y)$ and $V(x,y)$ satisfy Laplace's equation
    
* Let the choice of the coordinate system be such that in the $xy$ plane, the cylinders are represented by two circles of radius $r$, the first centered at $x=c, y=0$, and the second at $x=-(r^2/c), y=0$

    ><img src="figs/gmath1_06.jpg" width="400">
    
    Here, $c>r$ is a positive parameter which determines the distance between the circles. With our parametrization, the requirement that the distance between the centers of two circles is larger than the diameter of a cirle is automatically satisfied
    
    >$\begin{align*}
     (c -r)^2 &> 0 \\
     &\Downarrow \\ 
     c +\frac{r^2}{c} &> 2r
    \end{align*}$
    
    It is not difficult to see that the left side circle contains the origin in its interior
    
    >$\displaystyle \frac{r^2}{c} = r \left(\frac{r}{c}\right) \leq r$
    
* We seek a function $F(z)$ which is analytic outside the two circles and whose real part is constant on the circumference of each of the two circles. Performing the inversion

    >$\displaystyle z'=\frac{1}{z}$
    
    one obtain in the $z'$ plane two cencentric circles with radii
    
    >$\begin{align*}
 \left| z +\frac{r^2}{c} \right| &\geq r \\ 
 &\Downarrow \\ 
 \left| \frac{1}{z'} +\frac{r^2}{c} \right| &\geq r \\
 &\Downarrow \\ 
 \left| z' -\frac{c}{c^2 -r^2} \right| &\leq \frac{c^2}{r(c^2 -r^2)} =r_2
\end{align*}$

    and
    
    >$\begin{align*}
 \left| z -c \right| &\geq r \\ 
 &\Downarrow \\ 
 \left| \frac{1}{z'} -c \right| &\geq r \\
 &\Downarrow \\ 
 \left| z' -\frac{c}{c^2 -r^2} \right| &\geq \frac{r}{(c^2 -r^2)} = r_1
\end{align*}$
    
    ><img src="figs/gmath1_07.jpg" width="320">
    
    The outside of the circles in the $z$ plane is mapped onto the annular region
    
    >$\displaystyle r_1 \leq \left| z' -\frac{c}{c^2 -r^2} \right| \leq r_2$
    
    in the $z'$ plane. Set, for simplicity of notion, $w=z' -\frac{c}{c^2 -r^2}$. The function satisfying our requirement is
    
    >$\displaystyle F(w) = a\mathrm{Ln}\, w + b$, where $a$ and $b$ are real constants
    
    where $\mathrm{Ln}\, w$ is analytic in the annular region except for $\mathrm{arg}\,w=\pm\pi$, where it is double-valued.
    The real part of $F(w)$ is
    
    >$\displaystyle \mathrm{Re}\, F(w) = a\ln |w| +b$
    
    and the constants $a$ and $b$ are determined from the boundary conditions
    
    >$\mathrm{Re}\,F(w)=
\begin{cases}
 \;\;U_1 & \text{ for } |w|=r_1 \\ 
 \;\;U_2 & \text{ for } |w|=r_2 
\end{cases}$

    $U_1$ and $U_2$ denote the potentials of the first and of the second cylinders, respectively. Hence
    
    >$\begin{align*}
       a &= \frac{U_1 -U_2}{\ln\left(r_1/r_2 \right)}\\ 
       b &= \frac{U_2\ln r_1 -U_1 \ln r_2}{\ln\left(r_1/r_2 \right)}
      \end{align*}$
      
    After performing simple algebra, we obtain the solution of the problem
    
    >$\displaystyle U(x,y)= a \ln \,\left| \frac{1}{x +iy} -\frac{c}{c^2 -r^2} \right| +b$

### 11. The Cauchy Theorem

* A region is said to be **simply connected** if it is such that all closed paths within it contain only points that belong to the region (no holes). Otherwise, the region is said to be **multiply connected**

    ><img src="figs/gmath1_08.jpg" width="320">

* An arbitrary function of a complex variable will not possess a sinlge-valued primitive function, and the result of the contour integration will almost invariably depend upon the path linking the two end points. On the other hand, analytic functions have a remarkable property. Within simply-connected regions, they always possess single-valued primitive functions, so that the result of their integration is independent of the choice of the path

**Theorem** $\text{ }$ Let $\mathrm{C}$ denote a piecewise, regular closed curve in the complex plane and let $f(z)$ be analytic on $\mathrm{C}$ and within the whole region enclosed by $\mathrm{C}$. Then

>$\displaystyle \int_\mathrm{C} f(z)\, dz = 0$

* A consequence of the proceeding theorem is the following: Suppose that $\mathrm{C}_1$ and $\mathrm{C}_2$ are two curves that have the same end points, and suppose further that $f(z)$ has no singularities in the region enclosed by $\mathrm{C}_1$ and $\mathrm{C}_2$. Then $\mathrm{C}_1 +(-\mathrm{C}_2)$ is a closed contour, and therefore, using Cauchy's theorem, we find

>$\begin{align*}
 0 &= \int_{\mathrm{C}_1+(-\mathrm{C}_2)} f(z)\, dz\\ 
 &= \int_{\mathrm{C}_1} f(z)\, dz +\int_{-\mathrm{C}_2} f(z)\, dz\\ 
 &= \int_{\mathrm{C}_1} f(z)\, dz -\int_{\mathrm{C}_2} f(z)\, dz\\ 
 &\Downarrow \\
 \int_{\mathrm{C}_1} f(z)\, dz &= \int_{\mathrm{C}_2} f(z)\, dz
\end{align*}$

### 12. Cauch's Integral Representation

**Theorem** $\text{ }$ Let $f(z)$ be analytic throughout a simply-connected region $\mathrm{R}$. If $\mathrm{C}$ is a closed, piecewise, regular curve within $\mathrm{R}$, and $z$ a point not on $\mathrm{C}$, then

>$\displaystyle
\frac{1}{2\pi i} \int_\mathrm{C} \frac{f(z')}{z' -z} \,dz' =
\begin{cases}
 f(z) & \text{ if } z \text{ is interior to } \mathrm{C} \\ 
 \;\;\,0 & \text{ if } z \text{ is exterior to } \mathrm{C} 
\end{cases}
$

where the integration along $\mathrm{C}$ is taken in the counterclockwise direction

**Proof** $\text{ }$ Naturally, if $z$ is exterior to $\mathrm{C}$, the integrand is analytic and the RHS vanishes.

Consider the function

>$\displaystyle \frac{f(z') -f(z)}{z' -z}$

Since $f(z)$ is continuous, we have

>$|f(z') -f(z)| < \epsilon$

whenever

>$|z' -z| < \delta(\epsilon)$

Let $\Gamma$ be a circle in $\mathrm{R}$ centered at $z$ and of radius $r <\delta(\epsilon)$. Suppose that we integrate along $\Gamma$ in the counterclockwise direction. Then the parametric equation for $\Gamma$ is

>$z'(\theta)=z +re^{i\theta}$, $0 \leq \theta \leq 2\pi$

and for $z'$ on $\Gamma$, we have

>$\displaystyle \left| \frac{f(z') -f(z)}{z' -z} \right| < \frac{\epsilon}{r}$

Hence, using Darboux's inequality

>$\displaystyle \left| \int_\Gamma \frac{f(z') -f(z)}{z' -z} \, dz' \right| \leq \frac{\epsilon}{r} 2\pi r =2\pi \epsilon$

Therefore, in the limit $\epsilon \to 0$, the RHS tends to zero, and hence

>$\displaystyle \int_\Gamma \frac{f(z') -f(z)}{z' -z} \, dz' = 0$ $\;\;(*)$

The equation $(*)$ yields

>$\displaystyle \int_\Gamma \frac{f(z')}{z' -z} \,dz' = f(z)\int_\Gamma \frac{dz'}{z' -z} 
 = f(z)\int_0^{2\pi} \frac{ir e^{i\theta}}{re^{i\theta}}\,d\theta =2\pi i f(z)$
 
The function $f(z')/(z' -z)$ is an analytic function of $z'$, except in general at the point $z'=z$; therefore the boundary circle $\Gamma$ can be deformed into an arbitrary closed path $\mathrm{C}$ within $\mathrm{R}$, encircling the point $z$ and not passing through it. Hence, we finally obtain the general form of Cauchy's integral formula

>$\displaystyle f(z) =\frac{1}{2\pi i} \int_\mathrm{C} \frac{f(z')}{z' -z} \,dz'$

where $z$ is an interior point of an otherwise arbitrary curve $\mathrm{C}$. This equation is the so-called integral representation of a function. One says that one has established an integral representation for a function $f(z)$ if one has succeeded in writing down an integral of the form

>$\displaystyle f(z=\int_\mathrm{C} K(z,z')\, dz'$

where the function $K(z,z')$, which depends on both the parameter $z$ and the integration variable $z'$, is called the **kernel** of the integral representation. When the integrand is an analytic function and the integral is or may be reduced to a contour integral, the Cauchy theorem offers the possibility, by a judicious deformation of the integration path, to derive approximate expressions for the function

* The Cauchy's integral formula expresses completely the value of an anlaytic function at any point $z$ within a contour $\mathrm{C}$, once its values on the boundary curve have been specified - the nonlocal implications of the differentiablity of a function with respect to a complex argument

* A function of a complex variable must satisfy a differential equation, namely, the Laplace equation, in order for it to have a derivative

### 13. The Derivatives of an Analytic Function

We shall demonstrate that any function $f(z)$ which can be represented as

>$\displaystyle f(z)=\frac{1}{2\pi i} \int_\mathrm{C} \frac{g(z')}{z'-z}\,dz'$ $\;\;(\circ)$

where $\mathrm{C}$ is a piecewise, regular curve of finite length (not necessarily closed) and $g(z)$ a function that is continuous on $\mathrm{C}$, is analytic at any point $z$ which does not lie on $\mathrm{C}$. To this end, let us consider the expression

>$\displaystyle \begin{align*}
 \Delta &\equiv \left| \frac{f(z+\Delta z) -f(z)}{\Delta z} -\frac{1}{2\pi i} \int_\mathrm{C} \frac{g(z')}{(z' -z)^2} \,dz' \right| \\ 
 &= \frac{1}{2\pi}\left| \int_\mathrm{C} \left[ \frac{1/\Delta z}{z'-z-\Delta z} -\frac{1/\Delta z}{z' -z} 
   -\frac{1}{(z' -z)^2} \right] \,g(z') \,dz'\right|\\ 
 &= \frac{1}{2\pi} \left| \Delta z  \int_\mathrm{C}  \frac{\,g(z')}{(z'-z-\Delta z) (z'-z)^2} \,dz'\right|\\ 
 &\Downarrow {\scriptstyle \text{Since } z \text{ is not on } \mathrm{C}, \text{ the integrand is bounded }} \\
 \Delta \to 0\;\; &\text{ as } \Delta z \to 0
\end{align*}$

This proves the differentiability of $f(z)$ and also demonstrates that

>$\displaystyle \frac{df(z)}{dz} = \frac{1}{2\pi i} \int_{\mathrm{C}} \frac{g(z')}{(z' -z)^2} \,dz'$

An analogous result holds for the $n$-th derivative of $f(z)$

>$\displaystyle \frac{d^nf(z)}{dz^n} = \frac{n!}{2\pi i} \int_{\mathrm{C}} \frac{g(z')}{(z' -z)^{n +1}} \,dz'$

* In particular , a function $f(z)$ that is analytic can be expressed in the region by Cauchy's integral formula, which has exactly the form of $(\circ)$; $\mathrm{C}$ is now an arbitrary closed contour encircling the point $z$ and $g(z) \equiv f(z)$

**Theorem 1** $\text{ }$ The derivatives of all order of an analytic functions are themselves analytic

**Theorem 2** $\text{ }$ Given an integral representation

>$\displaystyle f(z)=\int_\mathrm{C} K(z,z') g(z')\,dz'$, $\;\; z \in \mathrm{R}\;\;$ $(\diamond)$

the derivative of $f(z)$ is given by

>$\displaystyle \frac{df(z)}{dz}=\int_\mathrm{C} \frac{\partial K(z,z')}{\partial z} g(z')\,dz'$

provided the following conditions are fulfilled:

1. For $z \in \mathrm{R}$, $K(z,z')$ is an analytic function of $z$ for any $z'$ on the contour $\mathrm{C}$

2. For any $z \in \mathrm{R}$, $K(z,z')g(z')$ is a continuous function of $z'$

**proof** $\text{ }$ Because of 1, we can write

>$\displaystyle K(z,z')=\frac{1}{2\pi i} \int_{\mathrm{C}'} \frac{K(t, z')}{t-z'}\,dt$, $\;\;\mathrm{C}' \in \mathrm{R}$

Inserting this expression in $(\diamond)$ and interchanging the order of integrations, we obtain

>$\displaystyle f(z)=\frac{1}{2\pi i} \int_{\mathrm{C}'} \frac{\int_{\mathrm{C}} K(t,z') g(z')\,dz' }{t -z} \,dt$

Since the integral in the numerator is a continuous function of $t$, we have

>$\displaystyle \frac{df(z)}{dz}=\frac{1}{2\pi i} \int_{\mathrm{C}'} \frac{\int_{\mathrm{C}} K(t,z') g(z')\,dz' }{(t -z)^2} \,dt = \int_{\mathrm{C}} \left[ \frac{1}{2\pi i} \int_{\mathrm{C}'}\frac{K(t,z')}{(t -z)^2} \,dt\right]  g(z')\,dz'$

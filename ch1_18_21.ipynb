### 18. The Taylor Series

**Theorem** $\text{ }$ Let $f(z)$ be a function, analytic within and on a cirlce centered at $z=z_0$. The value of this function at any point $z$ within $\Gamma$ is given by the uniformly convergent power series

>$\displaystyle f(z)=\sum_{n=0}^\infty a_n (z -z_0)^n$

where

>$\displaystyle a_n=\frac{1}{n!} \left.\frac{d^n f(z)}{dz^n}\right|_{z=z_0} =\frac{1}{2\pi i} \int_\Gamma \frac{f(z')}{(z' -z_0)^{n+1}}\, dz'$

**proof** $\text{ }$ Since $f(z)$ is analytic, we have

>$\displaystyle f(z)=\frac{1}{2\pi i} \int_\Gamma \frac{f(z')}{z' -z} \,dz'$

We expand the denominator as

>$\begin{align*}
 \frac{1}{z' -z} &= \frac{1}{(z' -z_0) -(z -z_0)} \\ 
 &= \frac{1}{(z' -z_0)} \cdot  \frac{1}{\left( \displaystyle 1 -\frac{z -z_0}{z' -z_0} \right)}\\ 
 &= \frac{1}{(z' -z_0)} \sum_{n=0}^\infty \left( \frac{z -z_0}{z' -z_0} \right)^n, \;\;\; 
   \left| \frac{z -z_0}{z' -z_0} \right| < 1
\end{align*}$

The above geometric series is therefore uniformly convergent and we find

>$\displaystyle f(z)=\frac{1}{2\pi i} \sum_{n=0}^\infty (z -z_0)^n \int_\Gamma \frac{f(z')}{(z' -z_0)^{n+1}} \,dz'$

* What is the radius of convergence of the Taylor series? The answer is given immediately by inspection of Cauchy's formula 

    >$\displaystyle f(z)=\frac{1}{2\pi i} \int_\Gamma \frac{f(z')}{z' -z} \,dz'$

   on which the proof of the Taylor expansion rests. Indeed, this formula breaks down when $\Gamma$ goes through or encircles a singularity of $f(z)$. Therefore, we are led to the conclusion that the radius of convergence of the power series cannot be greater than the distance from the point $z=z_0$ to the nearest singularity of $f(z)$. As an example, the series $1+z+z^2+\cdots=1/(1-z)$ converges in the interior of the circle $|z|=1$
   
* We have demonstrated that an analytic function can be expanded in a Taylor series at any of its regular points. The converse is also true; i.e., a function that can be expanded in a power series

    >$\displaystyle f(z)=\sum_{n=0}^\infty a_n(z-z_0)^n$
    
    which is convergent in some neighborhood of the point $z=z_0$ is neccessarily analytic

### 19. Poisson's Integral Representation

Consider the Taylor series

>$\displaystyle f(z)=\sum_{n=0}^\infty a_n z^n$

where the complex constants $a_n$ are given by

>$\displaystyle a_n=\frac{1}{2\pi i} \int_\Gamma \frac{f(z')}{z'^{n+1}}\, dz'$ $(*)$

Equation $(*)$ is also meaningful for $n < 0$, but naturally $a_{-|n|}$ (and hence $\bar{a}_{-|n|}$) vanishes, since the integrand in $a_{-|n|}$ is analytic. Letting $z'=R e^{i\theta'}$ in $(*)$, and using the notation $f(R,\theta)=f(Re^{i\theta})$, one has

>$\displaystyle a_n=\frac{1}{2\pi R^n} \int_0^{2\pi} \left[ \mathrm{Re} f(R,\theta')
  +i\, \mathrm{Im} f(R,\theta') \right] e^{-in\theta'} \,d\theta'$
  
The equation $\bar{a}_{-|n|}=0$ can thus be written as

>$\begin{align*}
 \bar{a}_{-|n|} &= \frac{1}{2\pi R^{-|n|}} \int_0^{2\pi} \overline{ \left[ \mathrm{Re} f(R,\theta')
  +i\, \mathrm{Im} f(R,\theta')  \right]e^{i|n|\theta'}} \,d\theta'\\
  &= \frac{1}{2\pi R^{-|n|}} \int_0^{2\pi} \left[ \mathrm{Re} f(R,\theta')
  -i\, \mathrm{Im} f(R,\theta') \right] e^{-i|n|\theta'} \,d\theta' = 0\\ 
 &\Downarrow \\ 
 \int_0^{2\pi} &\mathrm{Re} f(R,\theta')\, e^{-in\theta'} d\theta'
  = i\int_0^{2\pi} \mathrm{Im} f(R,\theta')\, e^{-in\theta'} d\theta', \;\; (n>0) 
\end{align*}$

Hence

>$\displaystyle a_n=\frac{1}{\pi R^n} \int_0^{2\pi} \mathrm{Re} f(R,\theta') e^{-in\theta'} \,d\theta', \;\; (n>0)$

Setting $z=re^{i\theta}$ with $r<R$, we find

>$\begin{align*}
 f(r,\theta) &= \frac{1}{2\pi} \int_0^{2\pi} \mathrm{Re} f(R,\theta') \,d\theta'
  +\frac{i}{2\pi} \int_0^{2\pi} \mathrm{Im} f(R,\theta') \,d\theta'  \\
  &+\frac{1}{\pi} \sum_{n=1}^\infty \left( \frac{r}{R} \right)^n 
  \int_0^{2\pi} \mathrm{Re} f(R,\theta') e^{in(\theta -\theta')}\,d\theta'\\ 
 &= \frac{i}{2\pi} \int_0^{2\pi} \mathrm{Im} f(R,\theta') \,d\theta'
  +\frac{1}{2\pi} \int_0^{2\pi} \mathrm{Re} f(R,\theta') \left[ 1 + 2\sum_{n=1}^\infty \left(\frac{re^{i(\theta-\theta')}}{R} \right)^n \right]\,d\theta' \\
 &= \frac{i}{2\pi} \int_0^{2\pi} \mathrm{Im} f(R,\theta') \,d\theta'
  +\frac{1}{2\pi} \int_0^{2\pi} \mathrm{Re} f(R,\theta') \underbrace{\left[ \frac{Re^{i\theta'} 
  +re^{i\theta}}{Re^{i\theta'} -re^{i\theta}}\right]\left[ \frac{Re^{-i\theta'} 
  -re^{-i\theta}}{Re^{-i\theta'} -re^{-i\theta}}\right]}_{\frac{R^2 -r^2+i2rR\sin(\theta-\theta')}{R^2 -2rR \cos(\theta -\theta')+r^2}} \,d\theta' \\
 &\Downarrow \\
 \mathrm{Re} f(r,\theta) &= \frac{1}{2\pi} \int_0^{2\pi} \mathrm{Re} f(R,\theta') \frac{R^2 -r^2}{R^2 -2rR \cos(\theta -\theta')+r^2}\,d\theta', \;\; r<R
\end{align*} $

The preceding equation is known as Poisson's formula. Analogously, one can derive

>$\displaystyle \mathrm{Im} f(r,\theta) = \frac{1}{2\pi} \int_0^{2\pi} \mathrm{Im} f(R,\theta') \frac{R^2 -r^2}{R^2 -2rR \cos(\theta -\theta')+r^2}\,d\theta', \;\; r<R$

These equations are examples of the so-called Dirichlet principle applied to a circle, whereby the value of an harmonic function at an interior point of a closed curve can be determined, once the values of this function along the boundary are given

#### 20. The Laurent Series

If a function $f(z)$ is not analytic throughout the whole interior of a circle, but only throughout the annular region between two concentric circles $\Gamma_1$ and $\Gamma_2$, it is possible to generalize the Taylor expansion of $f(z)$, which now becomes an expansion in both positive and negative powers of $(z -z_0)$. Such an expansion is called a **Laurent series**

**Theorem** $\text{ }$ Let $f(z)$ be analytic in the annular region between and on two concentric circles $\Gamma_1$ and $\Gamma_2$ centered at $z=z_0$. The value of $f(z)$ at any point $z$ within the annular region is given by the uniformly convergent power series

>$\displaystyle f(z)=\sum_{n=-\infty}^{\infty} d_n (z -z_0)^n$ $\;\;(\circ)$

where

>$\displaystyle d_n=\frac{1}{2\pi i} \int_\mathrm{C} \frac{f(z')}{(z' -z_0)^{n+1}}\,dz'$

and $\mathrm{C}$ is any contour within the annular region encircling the point $z_0$

**proof** $\text{ }$ We draw a small circle $\gamma$ centered at $z$ and contained entirely within the annular region of analyticity of $f(z)$. Consider the closed contour $\mathrm{C}'$

><img src="figs/gmath1_09.jpg" width="320">

Acoording to Cauchy's theorem

>$\displaystyle \int_{\mathrm{C}'} \frac{f(z')}{z' -z} \,dz' = 0$ $\;\;(*)$

since the integrand is analytic in the region enclosed by $\mathrm{C}'$. The parallel straight segments of the path can lie arbitrarily close to each other and thus give contributions equal in magnitude but of opposite sign; therefore, equation $(*)$ can be rewritten as

>$\begin{align*}
 \underbrace{\int_{\gamma} \frac{f(z')}{z'-z} dz'}_{2\pi i f(z)} &= \int_{\Gamma_2} \frac{f(z')}{z'-z} dz' 
  -\int_{\Gamma_1} \frac{f(z')}{z'-z} dz'\\ 
 &\Downarrow \\ 
 f(z) &= \underbrace{\frac{1}{2\pi i} \int_{\Gamma_2} \frac{f(z')}{z'-z} dz'}_{\sum_{n=0}^\infty a_n (z -z_0)^n, \;\;a_n=\frac{1}{2\pi i} \int_{\Gamma_2} \frac{f(z')}{(z' -z_0)^{n+1}}dz'}  \\ 
 &+ \underbrace{\frac{1}{2\pi i} \int_{\Gamma_1} \underbrace{\frac{1}{(z-z_0) -(z' -z_0)}}_{\frac{1}{z -z_0} \sum_{n=1}^\infty \left( \frac{z' -z_0}{z -z_0} \right)^{n -1}} \, f(z')\, dz'}_{\sum_{n=1}^\infty \frac{b_n}{(z -z_0)^n},\;\;b_n=\frac{1}{2\pi i} \int_{\Gamma_1} (z' -z)^{n -1} f(z')\,dz' } \\
 \Downarrow \\
 f(z) &=\sum_{n=-\infty}^{\infty} d_n (z-z_0)^n, \;\;d_n=\frac{1}{2\pi i} \int_\mathrm{C} \frac{f(z')}{(z' -z_0)^{n +1}}\,dz'
\end{align*}$

The sum $(\circ)$ converges uniformly to $f(z)$ for

>$\displaystyle R_1 < |z -z_0| < R_2$

$R_1$ and $R_2$ denoting respectively the radii of the circles $\Gamma_1$ and $\Gamma_2$, since this is the condition for the simultaneous convergence of the expansions. Again, the annular region throughout which the series converges can be enlarged until the first singular point of the function $f(z)$ is reached

### 21. Zeros and Isolated Singular Points of Analytic Functions



#### 21.1 Zeros

If a function $f(z)$ vanishes at a point $z=z_0$, this point is called a **zero** of $f(z)$. A function is said to have a zero of order $n$ at $z=z_0$ if

>$\displaystyle f(z_0)=\left.\frac{df(z)}{dz}\right|_{z=z_0} =\cdots=\left.\frac{d^{n-1}f(z)}{dz^{n-1}}\right|_{z=z_0}=0$

but

>$\displaystyle \left.\frac{d^n f(z)}{dz^n}\right|_{z=z_0} \neq 0$

Then the first $n$ coefficients in the Taylor expansion of $f(z)$ about $z=z_0$ vanish so that

>$\begin{align*}
 f(z) &= a_n(z -z_0)^n +a_{n+1}(z -z_0)^{n +1} + \cdots\\ 
 &= (z -z_0)^n \sum_{k=0}^\infty a_{n +k} (z -z_0)^k\\ 
 &= (z -z_0)^n h(z)
\end{align*}$

Here $h(z)$ is analytic and nonvanishing at $z=z_0$. Since $h(z)$ is continuous, it must also differ from zero in some finite neighborhood of $z=z_0$, and the same is true for $f(z)$. This implies that the zeros of an analytic function are isolated. Thus the set of zeros of an analytic function cannot have an accumulation point except when $f(z)\equiv 0$ 

#### 21.2 Isolated Singular Points

In order to give a classification of isolated singular points of analytic functions, it is very convenient to consider the Laurent expansion of the function $f(z)$ about a point $z_0$

>$\displaystyle f(z)=\sum_{n=0}^\infty a_n (z -z_0)^n+\frac{b_1}{z-z_0}+\frac{b_2}{(z-z_0)^2}+\cdots$

Suppose that $f(z)$ has an isolated singularity at the point $z=z_0$ and is analytic within a circle centered at this point. Then the annular region where the preceding expansion converges will reduce to the whole interior of the circle with the point $z=z_0$ taken out

It is clear that $f(z)$ may have a singularity at $z=z_0$ if at least one of the $b_j$ is not equal to zero. If the coefficient $b_n$ does not vanish, but all higher coefficients $b_j$ dp vanish

>$\displaystyle b_{n+1}=b_{n+2}=\cdots=0$

then the function $f(z)$ is said to have a **pole of order** $n$ at $z=z_0$. The sum

>$\displaystyle \frac{b_1}{z-z_0} +\frac{b_2}{(z-z_0)^2}+\cdots+\frac{b_n}{(z-z_0)^n}$

is then referred to as the **principal part** of $f(z)$ at $z=z_0$. If $b_1\neq 0$, but $b_2=b_3=\cdots=0$, the function is said to have a **simple pole**

* A function that is analytic in a region of the complex plane, except at a set of points of the region where the function has poles, is called a **meromorphic** function in this region

* When there is an infinite number of coefficients that do not vanish in the Laurent expansion of $f(z)$ about $z_0$, the function is said to have an **isolated essential singularity** at $z=z_0$

**Theorem** $\text{ }$ If a function $f(z)$ has an isolated essential singularity at a point $z=z_0$, then for arbitrary positive numbers $\epsilon$ and $\delta$ and for **any** complex number $a$, one has

>$\displaystyle |f(z) -a| < \epsilon$

for some point $z$ satisfying $|z -z_0| <\delta$

Expressed differently: In an arbitrary neighborhood of an essential singularity, a function oscillates so rapidly that it comes arbitrarily close to any possible complex number

**proof** $\text{ }$ We first convince ourselves that the sum of an infinite number of singular terms in the Laurent expansion cannot be bounded in the neigborhood of $z=z_0$. Applying the Darboux inequality to the integral

>$\displaystyle b_n=\frac{1}{2\pi i} \int_\Gamma (z' -z)^{n-1} f(z')\,dz'$

where $\Gamma$ is a circle of radius $r$ centered at $z=z_0$, we get for $n>1$, 

>$\displaystyle |b_n| \leq \left.\frac{r^{n-1}}{2\pi} \mathrm{max} |f(z)| \cdot 2\pi r = r^n \mathrm{max} |f(z)| \right|_{r\to 0} \to 0$

Thus, $b_n=0$ for $n>1$, in contradiction with the assumption that there exists an essential singularity at $z=z_0$. Hence $f(z)$ cannot be bounded

Take an arbitrary complex number $a$. The point $z=z_0$ is or is not an accumulation point of the zeros of the function $f(z) -a$. If it is, then (since in an arbitrary neighborhood of $z_0$, one has at least one point where $f(z) -a=0$), the theorem follows immediately. If $z_0$ is not an accumulation point of the zeros of $f(z) -a$, then for some $\eta$, $f(z)-a \neq 0$, provided

>$\displaystyle |z -z_0| < \eta$

Consider the function

>$\displaystyle g(z)=\frac{1}{f(z) -a}$

which is well defined in the region $|z -z_0| < \eta$ enclosing the point $z_0$. Since

>$\displaystyle f(z)=\frac{1}{g(z)} +a$

the function $g(z)$ must also have an essential singularity at $z=z_0$, since otherwise $f(z)$ would either be analytic or it would have a pole at this point. Therefore, $g(z)$ cannot be bounded within any circle $|z-z_0|<\delta$, and in particular there must be a point where

>$\displaystyle |g(z)|>\frac{1}{\epsilon} \Leftrightarrow |f(z) -a| <\epsilon$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

x = lambda r, t: np.outer(r, np.cos(t))
y = lambda r, t: np.outer(r, np.sin(t))
z = lambda r, t: np.outer(np.ones_like(r), t)

fig = plt.figure(figsize=[8,8])
ax = fig.add_subplot(111, projection='3d')

#----- 

t = np.linspace(-np.pi, np.pi, 50)
r = np.linspace(1, 10, 9)
for rr in r:
    ax.plot(x(rr, t)[0], y(rr, t)[0], z(rr, t)[0], color='k', linewidth=.8, alpha=1)
ax.plot(x([0, 10],-np.pi)[:, 0], y([0, 10],-np.pi)[:, 0], z([0, 10],-np.pi)[:, 0], color='r', linewidth=2.0)
ax.plot(x([0, 10], np.pi)[:, 0], y([0, 10], np.pi)[:, 0], z([0, 10], np.pi)[:, 0], color='r', linewidth=2.0)

#-----

t = np.linspace(-3*np.pi, 3*np.pi, 150)
r = np.linspace(0, 10, 5)
ax.plot_surface(x(r, t), y(r, t), z(r, t), antialiased=True, shade=False, rstride=1, cstride=1, 
                cmap=cm.rainbow_r, alpha=0.65)               

#-----

ax.view_init(elev=30, azim=-115)

ax.set_xticks([-10,-5, 0, 5, 10])
ax.set_yticks([-10,-5, 0, 5, 10])
ax.set_zticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi])
ax.set_zticklabels([r'$-2\pi$', r'$-\pi$', '0', r'$\pi$', r'$2\pi$'])

ax.set_xlabel(r'$\mathrm{Re} z$')
ax.set_ylabel(r'$\mathrm{Im} z$')
ax.set_zlabel(r'$\mathrm{Im}\, f_n, \;n=-1,0,1$')

fig.savefig('Riemann_Surface_Log.svg', bbox_inches='tight', transparent=True)
plt.show()
```

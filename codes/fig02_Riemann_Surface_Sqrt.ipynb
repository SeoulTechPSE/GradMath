```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

x = lambda r, t: np.outer(r, np.cos(t))
y = lambda r, t: np.outer(r, np.sin(t))
z = lambda r, t: np.outer(np.sqrt(r), np.cos(t/2))

fig = plt.figure(figsize=[8,8])
ax = fig.add_subplot(111, projection='3d')

#----- 

t = np.linspace(0, 4*np.pi, 50)
r = np.linspace(0, 10, 10)
for rr in r:
    ax.plot(x(rr, t)[0], y(rr, t)[0], z(rr, t)[0], color='k', linewidth=.8, alpha=1)

r1 = np.linspace(0, 10, 50)  
ax.plot(x(r1, 0)[:, 0], y(r1, 0)[:, 0], z(r1, 0)[:, 0], color='r', linewidth=2.0)
ax.plot(x(r1, 2*np.pi)[:, 0], y(r1, 2*np.pi)[:, 0], z(r1, 2*np.pi)[:, 0], color='r', linewidth=2.0)

#-----

t = np.linspace(0, 4*np.pi, 50)
r = np.linspace(0, 10, 5)
ax.plot_surface(x(r,t), y(r,t), z(r,t), antialiased=True, shade=False, rstride=1, cstride=1, 
                cmap=cm.rainbow_r, alpha=0.65)               

#-----

ax.view_init(elev=25, azim=-100)

ax.set_xticks([-10,-5, 0, 5, 10])
ax.set_yticks([-10,-5, 0, 5, 10])
ax.set_zticks([-6,-3, 0, 3, 6])

ax.set_xlabel(r'$\mathrm{Re}\, z$')
ax.set_ylabel(r'$\mathrm{Im}\, z$')
ax.set_zlabel(r'$\mathrm{Re}\, z^{1/2}$')

fig.savefig('Riemann_Surface_Sqrt.svg', bbox_inches='tight', transparent=True)
plt.show()
```

## The Theory of Analytical Functions 

### 1. Elementary Notions of Set Theory and Analysis 

#### 1.1 Sets

* The notion of a **set** is basic to all of modern mathematics. We shall mean by a set a collection of objects, hereafter called **elements** of the set
* For example, the integers $1, 2, 3, \cdots, 99, 100$ form a set of $100$ elements. Another example of a set is given by the collection of all points on a line segment; here the number of elements is clearly infinite

#### 1.2 Some Notations of Set Theory

* We shall usually denote sets by capital letters; e.g. $\mathrm{A}$, $\mathrm{B}$. Sometimes, however, other symbols will also be used

> $(a,b)$ will denote the set of real numbers satisfying the inequality $a < x < b$
    
* The symbol $\in$ will frequently be used. It is an abbreviation for *belongs to*

> $x \in (a,b)$
> $a \in \mathrm{S} $ should be read *$a$ belongs to the set $\mathrm{S}$*
    
* When $\mathrm{A}$ and $\mathrm{B}$ are identical,

> $\mathrm{A}=\mathrm{B}$
    
* When each element of $\mathrm{A}$ is necessarily also an element of $\mathrm{B}$,

> $\mathrm{A} \subset \mathrm{B}$ or $\mathrm{B} \supset \mathrm{A},\;\;\;$ $\mathrm{A}$ is a **a subset** of $\mathrm{B}$
    
* The set that contains all elements of $\mathrm{A}$ and all elements of $\mathrm{B}$, but counted only once, is called the **sum** or the **union** of $\mathrm{A}$ and $\mathrm{B}$ and is denoted by

> $\mathrm{A} +\mathrm{B}$ or $\mathrm{A} \cup \mathrm{B}$
    
* The set of all elements common to both $\mathrm{A}$ and $\mathrm{B}$ is called the **intersection** or sometimes the **product**, of $\mathrm{A}$ and $\mathrm{B}$ and is denoted by

> $\mathrm{A} \cap \mathrm{B}$
    
* The set of all elements of $\mathrm{A}$ that are included in $\mathrm{B}$ is called the **difference** between $\mathrm{A}$ and $\mathrm{B}$ and is denoted by

> $\mathrm{A} - \mathrm{B}$ 
    
* It is convenient to the notion of an empty set. It plays the role of the number $0$ in algebra and it is denoted by $\phi$ 

#### 1.3 Sets of Geometrical Points

* The most straightforward classification of sets consists in distinguishing between finite sets and infinite sets

* A more subtle classification consists in distinguishing among infinite sets between enumerable and nonenumerable ones

    * A set is called **enumerable** if it is possible to establish a one-to-one correspondence between each of its elements and the set of natural numbers $1, 2, 3, \cdots$. Otherwise a set is called **nonenumerable**
    
    * Among infinite sets, the set of all integers, is obviously ennumerable, as is the set of all even (or odd) integers, or the set of all prime numbers
    
    * On the other hand, the set of all points of a line segment is nonenumerable. Other examples of nonenumerable sets are the interior of a closed curve in a plane and the interior of a closed surface in space
    
* By a **neighborhood** of a given point $p$, we shall mean a set of points $p'$ satisfying

>$\rho(p,p') < R $ where $\rho(p,p')$ denote the distance between $p$ and $p'$ and $R$ is an arbitrary positive number

* Using the concept of a neighborhood of a point, we can give a classification of the point sets. Consider a set $\mathrm{S}$ of geometrical points. 

    * A point $p \in \mathrm{S}$ is called an **isolated** point of the set if there exists a neighborhood of $p$  which does not contain any other point belonging to $\mathrm{S}$
    
    * A point, every neighborhood of which contains at least one element of $\mathrm{S}$, which is not identical with the point itself, is called an **accumulation point**
    
    * If not only a given point but also all points of some neighborhood of $p$ belong to $\mathrm{S}$, then $p$ is called an **interior** point of $\mathrm{S}$
    
    * Every interior point of a set is an accumulation point. The converse is not true. Moreover, an acculmulation point of a set need not necessarily belong to the set
    
* A set is called an **open set** if all its points are interior points; a set is called a **closed set** if it contains all its accumulation points

    * One can distinguish between an open interval $(a,b)$ that does not contain the numbers $a$ and $b$ and a closed interval $[a,b]$
    
* A **region** is an open set, any two points of which can be connected by a continuous line that is contained entirely within the set 

#### 1.4 The Complex Plane

* A complex number $z$ is completely specified by a pair of real numbers $x$ and $y$

> $z=x+iy\;\;$ where $i^2=-1$

* The real numbers $x$ and $y$ are called, respectively, the **real** and **imaginary** parts of $z$ and are denoted by

> $\begin{align*}
 \mathrm{Re} z&= x \\ 
 \mathrm{Im} z&= y 
\end{align*}$

* Any complex number can be represented by a point in a plane, hereafter called the **complex plane**

><img src="figs/gmath1_01.jpg" width="300">

>$\begin{align*}
 x &= r\cos\theta \\ 
 y &= r\sin\theta \\ 
 & \Updownarrow \\
 z &= x +iy = r (\cos\theta +i\sin\theta)\\
 & \Updownarrow \\ 
 r =& \sqrt{x^2 +y^2} \\
 \theta =& \tan^{-1} \frac{y}{x}
\end{align*}$

* $r$ is called the **modulus** of $z$ and is denoted by $|z|$

* $\theta$ is called the **argument** of $z$ and is denoted by $\mathrm{arg} z$. In fact the polar coordinate $\theta$ is determined only up to an integer multiple of $2\pi$

>$\mathrm{arg} z = \theta +2\pi k, \;\;k=0, \pm 1, \pm 2, \cdots$

* Given two complex numbers $z_1$ and $z_2$,

>$\begin{align*}
 z_1 \cdot z_2 &= (x_1 +iy_1)(x_2 +iy_2) \\ 
 &= r_1 r_2 (\cos\theta_1 +i\sin\theta_1)(\cos\theta_2 +i\sin\theta_2)\\ 
 &= r_1 r_2 \left[ \cos(\theta_1 +\theta_2) +i\sin(\theta_1 +\theta_2) \right] \\
 &\Downarrow \\
 |z_1 \cdot z_2| &= |z_1| |z_2| \\
 \mathrm{arg} (z_1 \cdot z_2) &= \mathrm{arg}\, z_1 +\mathrm{arg}\, z_2
\end{align*}$

* The addition of complex numbers can be represented graphically as the familiar rule of vector addition

>$\begin{align*}
 z_1 + z_2 &= (x_1 +iy_1) +(x_2 +iy_2) \\ 
 &= (x_1 +x_2) +i(y_1 +y_2) 
\end{align*}$

* Since the sum of the lengths of two sides of a triangle is larger than the length of the third side, one immediately gets the **triangle inequalities**

>$\begin{align*}
 |z_1 + z_2| &\leq |z_1| +|z_2| \\ 
 |z_1 + z_2| &\geq |z_1| -|z_2| \\ 
 |z_1 + z_2| &\geq |z_2| -|z_1| 
\end{align*}$

* The numbers $a=\mathrm{Re}\,a +i\mathrm{Im}\,a$ and $\bar{a}=\mathrm{Re}\,a -i\mathrm{Im}\,a$ are called **complex conjugates** of each other and are represented by points whose positions are symmetrical with respect to the real axis. The following relation is evident

>$a \cdot \bar{a} =(\mathrm{Re}\,a)^2 +(\mathrm{Im}\,a)^2 = |a|^2$

#### 1.5 Functions

* Let a function $f(p)$ be defined in a region $\mathrm{R}$; it is irrelevant to our discussion whether $\mathrm{R}$ is a set of points located *on a line, in a plane, or in space*

* The function $f(p)$ is said to be **continuous** at a point $p \in \mathrm{R}$ if for an arbitrary number $\epsilon > 0$, one can find a number $\delta > 0$ such that, provided that $\rho(p,p') < \delta$, one has

>$|f(p) -f(p')| < \epsilon$

* Consider an infinite sequence of functions defined in $\mathrm{R}$
    
    > $f_1(p), f_2(p), \cdots, f_n(p),\cdots $ $(*)$ 
    
    One says that this sequence converges in $\mathrm{R}$ to the function $f(p)$, if for any point $p \in \mathrm{R}$ and for an arbitrary $\epsilon > 0$ one can a number $N=N(p,\epsilon)$ such that, provided $n > N$, one has
    
    >$|f(p) -f_n(p)| < \epsilon$
        
    If $N$ is independent of $p$ throughout the set, then we say that the sequence $(*)$ converges **uniformly** to $f(p)$ in $\mathrm{R}$ 
    
* **The Cauchy Criterion**

    A sequence of functions $f_n(p)$ ($n=1,2,\cdots$) converges uniformly on a set $\mathrm{R}$ if and only if for any $\epsilon >0$, one can find a real number $N$ independent of $p \in \mathrm{R}$ and such that, provided $n > m> N$, one has
    >$|f_n(p) -f_m(p)| < \epsilon$

* **The Weierstrass Criterion**

    The infinite series
    
    >$\displaystyle \sum_{k=1}^\infty f_k(p)$
    
    converges uniformly if $|f_n(p)| \leq a_n$ and if the series
    
    >$\displaystyle \sum_{k=1}^\infty a_n$
    
    is convergent
